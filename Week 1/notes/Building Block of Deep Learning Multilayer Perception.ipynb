## Multi-layer Perception

- Multi-layer perception (MLP) is the simplest artificla neural network and can have many hidden layers
- Each hidden layer allows us to learn new and more complex features
- MLP is the building block of all Deep Learning
- Deep Learning is a term used for neural networks that have at least two hidden layers
----
- There are several important design parameters of a MLP architecture:
    1. Number of hidden layers.
    2. Number of neurons in each hidden layer.
    3. Whether the layers are fully-connected or not? If yes, then known as a Fully-Connected Neural Network.
    4. The choice of non-linear activation function. Much of the power of MLP and neural networks comes from the activation function.
