## Output Layer and Loss Minimization


- similar to the logistic function and it converts real valued numbers to probabilities in the range from 0-1
- it is common to apply the softmax function in the final or the output layer of a neural network if we are performing a classification task
- this is done bc in the end, we need classification probabilities'
- after we have a classification model, we optimize it to minimize the discrepancy between the correct and predicted outputs for various individuals
- the optimization procedure changes the model parameters (thetas) to make correct and predicted outputs simialr on the training dataset
- this process of loss minimization is exactly the same that is employed while learning standard machine learning models

## Some commonly used loss function for Neural Networks


- Binary Cross-Entropy Loss: 
    - loss function of choice for two class or binary classification problems (i.e., loan approval prediction)
    - this loss funciton can easily be generalized to a multiclass problem (i.e. loan approval prediction + conditional approval loan)

- Mean Squared Error Loss:
    - minimizes squared distance between the predicted and the correct label
    - commonly employed if we have a regression task (we want to predict a real valued output from our model)
