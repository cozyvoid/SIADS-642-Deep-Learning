# The Rise of Deep Learning

## Deeping Learning in the broader context of "machine intelligence"



**Artificial Intelligence (AI)**: A set of techniques that permit computers to mimic human behavior/intelligence

**Machine Learning (ML)**: A set of statistical techniques that enable computers to improve at tasks with experience

**Deep Learning (DL)**: Extracting patterns from large amounts of data via artifical neural networks

## How is Deep Learning (DL) different than traditional Machine Learning (ML)?

**Machine Learning 101**: Predict whether an image contains a cat or not? (0/1 classification task)

- Step 1: Collect labeled images (y) containing examples of "cats" and "not cats".
- Step 2: Manually extract features from the images (x)
- Step 3: Build a classifier to predict y using x e.g., SVM, logistic regression, random forests, etc.

## DL vs ML: DL models can automatically learn relevant features from the data

- We don't know the right level of *abstraction* for the features to be helpful in a prediction task.
- Deep Learning models help us *automatically* learn features at different *levels of abstraction*.

## Why the resurgence of Deep Learning (Neural Networks) now?

Though, Neural Networks date back decades (at least late 1950's), their resurgence in teh last decade has been due to (3) main reasons:

1.  **Big Data**: Access to larger (usually "digital") datasets. Ability to collect and store large datasets.
2. **Hardware**: Graphical Processing Units (GPU) became commerically viable leading to massive parallelization.
3. **Software**: Improved techniques implemented in public domain freamworks e.g., PyTorch, TensorFlow, Keras, MxNet.
