### Computer Vision and Reasoning with Images

- The field of **Computer Vision deals with understanding images** among other things.
- It involves tasks such as **object recognition**, **scene segmentation**, and **scene understanding**, etc.
- The **input** to computer vision models are **raw images**:
    - Raw images are just a matrix of numbers representing the intensity levels of various spatial locations.

### NN's for Learning Visual Features for Computer Vision Tasks

- **Solution (I)**: Flatten the image and input it to a **fully connected** Feedforward Neural Network (FFNN).



- **Problem**: Flattening **loses** the important **spatial structure**.

### NN's for Learning Visual Features for Computer Vision Tasks

- **Key Idea**: *Use spatial structure*. Connect patches of input image pixels to neruons in hidden layer of the neural network.



### Neural for Learning Visual Features for Computer Vision

- We need to **weight these extracted patches** from the image to **highlight their relative importance**.
- We need to use several of these **"spatial filters"** and at **different resolutions**, e.g., some to extract highly local features and others to extract more global features.
- **Convolutional Neural Networks (CNN's)** formalize this idea of **weighted spatial filters**.
    - They are a popular neural network architecture to extract local features from images.
