## Controlling model complexity in Neural Networks

- *Overfitting* and *underfitting* can be a big problem in Neural Network training
- It is important to *control model complexity via regularization* just as in the case of standard machine learning methods.
- **Regularization** is important to *improve the generalizability* of the trained model to unseen/test data.
- There are three main regularization approaches specifically tuned to neural network training: **Dropout**, **Early Stopping**, and **Data Augmentation**.
***

### 1. Dropout

- During training randomly set *d%* of the activations to 0 where *d ~ 15-20*.
- Forces network to not rely on any one node while making predictions.
- Essentially Dropout is a way of performing *model averaging using neural networks*.



### 2. Early Stopping

- Stop training the neural network model before it has fully converged. Thus giving it no chance to overfit (to have higher test error compared to the training error).



### 3. Data Augmentation

- For certain data domains such as *image data* it might be possible to *augment the training data* to prevent overfitting.
    - E.g., transforming the input data via *rotation, cropping, mirroring, jittering* or any other **output label preserving transformations**, and adding those transformed examples to the training data.
    - The hope is that the model trained on this augmented dataset will learn to *generalize beyond the variation already present in the training data*.

Though, data augmentation has mostly been used for computer vision and image data, but it is conceivable to employ it for other domains, as well (like text or speech).
