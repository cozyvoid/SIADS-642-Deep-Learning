### 1. Which of the following is not true regarding Deep Learning (DL) and traditional Machine Learning (ML)?

- [False] DL and ML models have similar number of parameters in general.
- [True] DL models learn abstractions of input data at different levels.
- [True] DL models have many more parameters than ML models.
- [True] DL models typically require more computational resources than ML models.

The statement that is not true is:

**DL and ML models have similar number of parameters in general.**

Deep learning models usually have far more parameters than traditional machine learning models, which is one reason they often need more data and more computational power.

### 2. What is each node in the computation graph of a multi-layer perceptron (MLP) called?

- A neuron

### 3. What is the name of the general algorithm that is used to train neural network with gradient descent?

- Backpropagation

**Backpropagation** is the general algorithm used to compute gradients for training neural networks with gradient descent.

### 4. Which of these weight parameter initialization methods work best for deep neural networks?

- Xavier Initialization

**Xavier initialization** works much better for deep neural networks than plain random values or all zeros. Initializing all weights to zeros breaks symmetry, and very large values cause unstable activations and gradients.

### 5. What problem is Batch normalization (BN) supposed to address?

- Internal covariate shift

Batch normalization was introduced to address internal covariate shift, meaning *changes in the distribution of layer inputs* during training.

### 6. Will changing the color space of an image from RGB to grayscale and adding it to training set constitute as data augmentation? Assume that the output label to be predicted is whether the image contains a “cat” or not. Also, assume that the number of channels in the input image is unchanged after the transformation.

- Yes

That counts as data augmentation because you are creating a modified version of the original input while keeping the same label (“cat” or “not cat”). Converting RGB to grayscale changes the image appearance but usually does not change the semantic content relevant to the task.

The assumption that the number of channels stays unchanged means the transformed image can still be fed into the model in the same input shape, so it is a valid augmentation.
