# Convolutional Neural Networks (CNN's)

## Recap: Convolution Filter Applied to an Input Image



### Convolutional Neural Networks (CNN's)

- CNN's treat the **weights of the convolution filter** as **model parameters** and learn them from data.



## Convolution Layers 

- For each neuron (a, b) in the hidden layer:
    1. Take a patch from the input image with pixel values denoted by x.
    2. Compute a weighted sum and add a bias term. For example, for a 3 x 3 filter, we would have:
    $$

    \sum_{i = 1}^{3} \sum_{j = 1}^3 \theta_{ij} x_{i+a},_{j+b} + \theta_0

    $$
    3. Apply a non-linear activation function.

    > All steps in the computaion of a convolutional layer are the same as in a "fully-connected layer" except that convolutional layers *share their parameters spatially*.

### Zero-padding the input image

- It is common to zero-pad the input image before applying the filter in order to preserve the size of the input image spatially.
    - Common to zero-pad with size (f-1) / 2 where f is the size of the filter. So, a zero-padding of 1 and 2 for filters of sizes 3x3 and 5x5 respectively.




### Other important practical details of Convolutional Filters
- **Stride of a filter**: It is the number of pixels by which the filter moves between consecutive moves while sliding over an image.

### CNN's in Practice: Pooling Layer

- The activation maps generated by different convolution filters **can get really large spatially**.
- This happens more so for deep networks leading to a **significant incraese in b oth the number of parameters and the amount of computation**.
- A **Pooling layer** makes the representations smalelr by **downsampling the activation map** but still preserving structure.



### Pooling Layer

- **Two common types of pooling**: Max-pooling & Average Pooling *(becoming less common these days)*.
- Pooling is applied to each activation map independently.



### CNN's: Putting it all together

- CNN's summarized:
    - Sequence of ConvLayers with non-linearities and occasional pooling.
    - The final layer before the output prediction is usually a fully-connected layer
