# Feature Extraction Using a Convolution

## What is a convolution?

- It is an **operation used in image processing** for tasks such as **edge detection, blurring, sharpening**, etc.
- Basic idea:
    - Pick a s*s matrix $C$ of weights. (Typically s = 2, 3, 5, or a small number)
    - **Slide it over the input image** computing element-wise products of C with corresponding pixel values.
    - **Different convolutions** (size "s" and weights) **extract diffrerent types of "low-level" features** from an image.
    - As we will see soon, **Convolutional Neural Networks (CNN's)** learn the convolution weights automatically from the data.

## Summary

- Convolutional filters of **different shapes and weights** can be used to select "local" feature maps from images **at different scales**.
