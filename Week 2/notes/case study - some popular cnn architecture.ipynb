# Case Study: Some Popular CNN Architecture

### LeNet-5: CNN for Document Recognition

- One of the first examples of a successful practical application of a deep CNN.



**Architecture**: ConvLayer-Pool-ConLayer-Pool-FC-FC

*Conv filters were 5x5 with stride1*. Pooling Layers were 2x2 with stide 2.

### AlexNet: CNN for Image Classification

- *Major breakthrough* in application of deep learning to computer vision.
- **Achieved state-of-the-art accuracy on ImageNet in 2012**. **[15.3% error]**. Trained on GTX 580 GPU with just 3GB of memory!



- Had 8 Layers. *First CNN-based winner of ImageNet challenge*.
- *First ever to use ReLU activation*, dropout rate of 0.5, much data augmentation.
- The very first convolution filter was of size 11x11 with stride size of 4. 
- Several follow-up architectures showed improvements with *smaller first layer filters* and *deeper networks* compared to AlexNet. 

### VGGNet (Simonyann & Zisserman 2014)

- **Key Idea**: smaller (3x3 filters with stride 1) and deeper networks. [7.3% error on ImageNet].
- 16 layers. A total of ~138M parameters.
- **Rationale for smaller filters**: Stack of smaller  filters has the same receptive field as some large filters, *BUT* allows more non-linearities and has 
fewer parameters.



### ResNet [He et al. 2015]: Deep Residual Learning Network

- Deeper Networks help learn better representations and lead to improved accuracies. However, much *deeper networks are very hard to train*.
- [He et al. 2015] showed a 56 layer network having both higher train and test accuracies than a 20 layer network.
- *Solution*: use NN layers to fit resisual value instead of the actual desired mapping.
- *ResNet had 152(!) layers* and achieved an **error-rate of 3.57%** on ImageNet.



Instead of fitting H(x) as done by a normal CNN layer, ResNet fits the residual $$ F(x) = H(x) - x $$



### The current state of CNN research

- The trend in recent years has been to go deeper with smaller filteres, residuals connections, and several engineering modifcations for compuational efficiency.
- Very active anf fast-movign area of research!
