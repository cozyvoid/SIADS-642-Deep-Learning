# Training CNN's

### Training of CNN's

- Broadly, the training of CNN's proceeds in a similar fashion as discussed in Week 1 for feedfoward neural networks.
That is,

    - Incorporating *non-linearities*
    - *Regularization* via Dropout, early stopping, and data augmentation (a popular choice for CNN's)
    - *Backpropagation* to update the parameters to minimize the loss. *Batch Normalization to improve the stability of the training procedure*.
    - *Gradient-based optimization* based on variants such as Adam, Adagrad, RMSProp, etc.
