# Concept Quiz 3

### 1. Which of the following is not a desiderata for modeling sequence data?

- Handle variable length sequences
- **Handle missing data**
- Preserve ordering information
- Model long-range dependencies

**Handle missing data**

The usual desiderata for sequence models are to:

- handle variable-length sequences
- preserve ordering information
- model long-range dependencies

Handling missing data can matter in some applications, but it is not typically listed as a core desideratum of sequence modeling itself.

### 2. Which Recurrent Neural Net (RNN) architecture would you use to build a machine translation model?

- one-to-one
- one-to-many
- many-to-one
- **many-to-many**

**many-to-many**

Machine translation maps an input sequence of words in one language to an output sequence of words in another language, so it uses a many-to-many architecture.

### 3. Which of the following is not a weakness of RNNs?

- Vanishing gradient problem
- Exploding gradient problem
- **Model size does not increase for longer sequences**
- Difficult to parallelize

**Model size does not increase for longer sequences**

That is **not** a weakness of RNNs. It is actually an advantage: the same recurrent cell parameters are reused across time steps, so parameter count does not grow with sequence length.

The actual weaknesses listed are:

- Vanishing gradients
- Exploding gradients
- Difficult to parallelize

### 4. Long short-term memory (LSTM) networks are designed to deal with which of the following problems?

- **Vanishing gradient problem**
- Exploding gradient problem
- Gradient saturation
- Preserve ordering information of sequence data

**Vanishing gradient problem**

LSTMs were designed primarily to help RNNs retain information over longer time spans, which addresses the vanishing gradient problem.

### 5. Which of the following is not a solution to the vanishing gradient problem?

- Use of ReLU activation function
- Use LSTM
- Use GRU
- **Gradient Clipping**

**Gradient Clipping**

Gradient clipping is mainly used to control the **exploding gradient** problem, not vanishing gradients.

Methods that help with vanishing gradients include:

- ReLU activation
- LSTM
- GRU

### 6. What is the sequence-to-sequence model also known as?

- **Encoder-Decoder Architecture**
- GRU
- Attentive Models
- Recursive Models

**Encoder-Decoder Architecture**

A sequence-to-sequence model is also known as an **encoder-decoder architecture.**
