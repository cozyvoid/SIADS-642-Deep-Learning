# Bidirectional and Multi-Layer RNN's

### Unidirectional RNN's

**Task**: Sentiment Classification (positive/negative)

- RNN's process information **sequentially from left-to-right**.
- So, the hidden state representation of some word e.g. "but" **only contains information about its left context** "not bad", however it contains no information about its right content "good".
- For many language tasks, modeling both left and right contexts can help us learn better performing models.

### Multi-layer RNN's 

- RNN's unrolled over many timesteps are **already "deep" along one dimension**.
- We can make them "deep" along another dimension by **stacking multiple RNN's** aka *multi-layer RNN's* or *stacked RNN's*.
- Multi-layer RNN's allow one to compute **more complex representations** with **lower layers capturing more generic features** of the language and the **higher layers capturing more specific features** as CNN's do for images.



## Summary

- Bidirectional and Multi-layer RNN's can be built based on any RNN model e.g. vanilla RNN, LSTM, or GRU etc.
- **Most of the state-of-the-art** Natural Language Processing RNN architectures these days are **deep and bidirectional**.
- One such **popular deep bidirectional RNN** architecture for **learning language models is BERT** (Has 24 layers). Link to BERT paper: https://arxiv.org./abs/1810.04805
