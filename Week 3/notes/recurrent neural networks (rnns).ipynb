### Unrolling the RNN computational graph

- Some advantages of RNN's:
    - RNN's re-use the **same weight matrix** at each time-step and hence can easily process **varying length sequences**.
    - Relatedly, **model size does not increase** for longer input sequences.
    - Computation at a step "t" can use information from **several steps back**.



## Some shortcomings of RNN's

- recurrent computation is typically **slow and difficult to parallelize**.
- It is hard to access information from **many steps back** in practice.

- We will see some RNN architectures soon to help overcome this problem.

### Training of RNN's

Most of the training details ffor RNN's are the sam eas for other NN models that we have seen.

1. Minimize the loss function (typically cross-entropy for language modeling) over the entire dataset.
2. For computational efficiency use **minibatching** and **adaptive learning rates**.
3. **Optimize weights by backpropagation** which in the case of RNN's is called *Backpropagation through time* (BPTT)
