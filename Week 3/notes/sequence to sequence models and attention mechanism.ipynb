# Sequence to Sequence Models and Attention Mechanism

### RNN Architecture configurations



### Sequence to Sequence Models: Many-to-one + One-to-many

- Also called **"Encoder-Decoder"** architecture. Very popular for Neural Machine Translation (NMT) these days.
- Applications in many other language tasks e.g., *Summarization*, *Dialogue generation*, and *code generation*. 



### Case Study: Seq2Seq Models for NMT



### Seq2Seq Models for NMT: An Information Bottleneck

- The **entire information** about the input sequence is **condensed into the final hidden state of the Encoder RNN**.
- As a result, the **Decoder RNN just focuses on the final hidden state** while generating the decoded output.
- A solution to this information bottleneck can be found if we can have the **Decoder RNN "focus"** on **only the "relevant" parts of the Encoder RNN's hidden states**.
    - This is exactly what **Attention Mechanism** does!

### Attention 



### Benefits of Attention

- **Dramatically improves** the performance of NMT systems and **addresses the information bottleneck problem**.
    - It allows the decoder to focus on different parts of the input and "weight" them accordingly.
- Attention also helps with the **vanishing gradient problem** as it provides a shortcut to "distant" parts of the input.
- Provides **some interpretability** as we can see what the decoder was focusing on while generating a certain output.
