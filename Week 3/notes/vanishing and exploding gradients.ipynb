# Vanishing and Exploding Gradients

### Recap: Backpropagation through time (BPTT)



## Vanishing Gradient problem

- Recall that for a RNN: $ h_t = f(\theta_{hh}h_t{t-1}+ \theta_{xh}x_t)$

- Therefore: 



If $\theta_{hh}$ is small then the **gradient gets vanishingly small** if there are many successive multiplications.

### Vanishing and Exploding Gradients

- It can be show that if the largest eigenvalue of $\theta_{hh}$ **is less than 1**, then the gradient $ || \frac{\partial l^t}{\partial h_0} || $ will shrink exponentially leading to vanishing gradients.
- Similarly, if the largest eigenvalue **is greater than 1** we have exploding gradients. 

### Why is vanishing gradient a problem?

- Gradient signal from faraway gets lots as it becomes much smaller in magnitude compared to nearbvy signal.
- Therefore, model weights are only updated based on nearby information.
- Hence, RNN's are unable to capture long-term dependencies.

> Recall that modeling long-term text dependencies is one of the desiderata for many language modeling tasks.+

### Why is exploding gradient a problem?

- If the gradient becomes too big then the stochastic gradient descent (SGD) **updates can become too big**. 

$$
\theta^{t+1} = \theta^t - \eta \frac{\partial l}{\partial \theta^t}
$$

- This can lead to **instability in model training** requiring us to restart model training from the beginning.
- There is also the **possibility of reaching a bad parameter configuration** leading us to a **bad local optima**.

### Exploding gradient problemm solution: Gradient Clipping

- If the norm of the gradient becomes **greater than some threshold**, then **scale it down** before doing the gradient update.

- Broadly speaking, the idea behind **gradient clipping** is to take a step in the same direction, **but a smaller one**.

### Vanishing gradient problem solutions

- **Solution 1**: use **ReLU activation function**. It prevents gradients from shrinking when x > 0as it keeps all the gradient information. Hence, the gradients don't get progressively smaller.
- **Solution 2**: *Change the RNN architecture* to control what information passes through. 
    - This can be achieved via using **Gated Recurrent Cells e.g. LSTM, GRU**, etc. as opposed to a vanilla recurrent cell.
