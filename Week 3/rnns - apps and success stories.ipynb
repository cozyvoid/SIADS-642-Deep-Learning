# RNN's: Applications and Success Stories

### Sample RNN Applications

1. **Music Generation**: *Input*: sheet music, *Output*: Next character in sheet music.
2. **Sentiment Classification (or sentence classification)**: *Input*: words in the sentence, *Output*: classification label.
3. **Generating text**: Train RNN's on any text and tehn generate text in that style. Let's see some interesting examples!

### Biggest Success Story: Machine Translation (MT)

- Neural Machine Translation (NMT) gave the **most significant improvements in translation quality** with the *least amount of hand-engineering*.
    - Previous MT systems required lots of hands-engineering and consisted of several modules cobbled together
- In 2016 Google Translate switched to NMT!
