# Sequence Modeling

### Modeling Language

Lots of Practical Applications

    - Machine Translations
    - Automatic Speech Recognition



- Can we model **sequences** (of sets of words or acoustic waveforms etc. using the modeling approaches that we have learned till now?)

- Let's consider the following task: Predict the next word in a sentence.



Such prediction tasks are central to many practical language technologies and require a **language model (LM)**.

**Language Model**: a probability distribution $P(w_1, w_2,..., w_n)$ over a sequence of n words $w_1, w_2,..., w_n$

### Potential Solution I: A fixed window-based approach

- Take a **fixed-size word window (w)** around the word and use it to predict the word.

Sample Sentence:
*This afternoon we went for a stroll in the woods.*

- If w = 2, in the above sentence we would use the words **"in the"** to predict the word **"woods"**.

### The need to model Long-range dependencies

- The fixed-window approach can work well in certain cases but it is unable to model long-range dependencies in language.
- For example, consider the sentence:

**Michigan** is a state in the midwestern region of USA and shares a border with Canada. Its most populous city is **____. (Detroit)**


The word **"Michigan"** contains a strong cue to predict the word **"Detroit"** in the above sentence, but we need information from the *distant* past to predict it.

### Potential Solution II: Use set of counts from the entire sequence

*This afternoon we went for a stroll in the woods.*

- Use **counts** of words in the sequence to predict the next word e.g., *This = 1, afternoon = 1,...*
- Also known as a **Bag of words** approach.
- However, counts don't preserve the important ordering information.
- For example, the following sentences have exact same bag of words representation but contain totally different information:

    - *"Though the movie received bad reviews, I thought it was good."* &
    - *"Though the movie received good reviews, I thought it was bad."*

### Potential Solution III: Use "really big" fixed word windows

*This afternoon we went for a stroll in the woods.*

- Use all the words *"This"*, *"afternoon"*, *"we"*, "went", *"for"*, *"a"*, *"stroll"*, *"in"*, *"the"* to predict the word "woods".
- However, **big word windows** are computationally inefficient to compute and stor and also lead to an explosion in the number of model parameters.

## Desiderata for Sequence Modeling

1. Handle **variable-length** sequences.
2. Track and model **long-range** text dependencies.
3. Preserve text **ordering** information.
4. Do all of the above without blowing-up the parameter space and computational complexity of the model.

*We will study how **Recurrent Neural Networks** (RNN's) allow us to model sequence data while preserving the flexibilities of the NN modeling approach.*
