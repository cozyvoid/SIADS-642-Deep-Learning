### AutoML: Automated Machine Learning

- Learn to Learn (Zoph and Le 2017)
- Uses a RNN to **generate model descriptions** and then trains then mdoel with Reinforcement Learning to **search over the different neural architectures**.



### AutoML: RNN Model Controller



At each step: 1) the controller samples a bran new "child" network, 2) computes its prediction accuracy, and then 3) updates the RNN controller based on the accuracy of the "child" network.
