# Generative Adversarial Nets (GAN's)

### Generative Modeling

- **Recall**: The goal of generative modeling is to learn a model that approximates the distribution of the training data. That is, learn $ P_{model}^{(x)} \approx P_{training - data}^{(x)} $
- Two modeling strategies:
    1. **Directly model the density function** $ P_{model}^{(x)} $: Optimize it directly (or optimize a lower bound) e.g. PixelICNN, PixelRNN, or Variational Autoencoders (VAE).
    2. Implicity model the density function: Useful, if we just want to sample from $ p_{model}^{(x)} $. e.g. Generative Adversarial Network (GAN) and variants.

- GAN's introduce the new idea of "adversarial learning" that has found widespread use in DEeep Learning.

### Generative Adversial Nets (GAN's): Introduction

- Proposed by (Goodfellow et. al. 2014) and have become a propular model of choice for generative modeling.
- Incorporates ideas from **game-theory** to generate samples from the underlying density.
    - Models the task of generation from the training distribution as a **two-person game**.
- Hence, GAN's don't assume any explicit function form for the density function $ p_{model}^{(x)} $

### GAN's: Key Idea

- **Problem**: We want to sample from a complex distribution, but can't do it directly.
- **Solution**: Sample from a simple distribution e.g., random noise and then learn a transformation to the training distribution.
- The *complex transformation* from the *random noise to the training distribution* is represented using a *neural network!*

### GAN's: Architectural Details

- GAN's are a way to make a generative model by having *two neural networks compete with each other*.



### Training GAN's: Two player minimum

- **Generator (Neural) Network**: try to fool the discrimator by generating real-looking images.
- **Discrimator (Neural) Network**: Try to distinguish between real and fake images.



- *Discriminator* $(\theta_d)$ wants to *maximize the objective* s.t. $ D(x) $ is close to 1 (real) and $D(G(z))$ is close to 0 (fake).
- *Generator* $(\theta_g)$ wants to *minimize the objective* s.t. D(G(z)) is close to 1.

### Generting new data with GAN

- After training the GAN to optimize the parameters $ \theta_g $, $\theta_d $ use the **generator network** to generate new data.
