# Generative Models

## Recap: Supervised  vs Unsupervised Learning

| Supervised | Unsupervised|
|:---:|:---:|
|Input Data: $ (x, y) $ | Input Data: $(x)$ |
| $x$ = features, $y$ = labels | $x$ = features **(no labels!)** |
|Goal: Learning a mapping function $ x → y $ | Goal: Discover some underlying hidden structure in the data. |
|Example: All the classification and regression tasks that we have solved using FFNN's, CNN's, and RNN's in this course. | Example: Clustering, density estimation, feature learning, dimensionality reduction, etc. | 

### Generative Models

- **Goals**: Take as input training samples from some distribution and learn a model that  represents that distribution. In other words, learn $ P_{model}^{(x)} \approx P_{training - data}^ {(x)}$
- Hence allowing us to generate new samples from a given data distribution.
- As wel will see, this has a lot of important practical applications.
- The samples from input training data can be used for density estimation or sample generation grom the underlying distribution.



### Why Generative Models?

1. Uncover the underlying latent distributions of data and use it to *construct fair and represenative datasets*.
2. The *latent representations* inferred from generative models can be used as *features in a classifier*.
3. The generated data samples can be used to *augment the data* used for simulation and planning of robotic and autonomous vehicles.
4. Can be used to generate *realistic samples of artwork*.
