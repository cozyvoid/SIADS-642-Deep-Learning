# Limitations of Deep Neural Networks

### Power of Neural Networks

- **Universal Approximation Theorem**: A feedforward network with a single layer is sufficient to approximate, to any artbitrary precision, any continuous function. [Hornik at al. 1989]
- **Caveats**:
    - The *number of hidden units* required in that hidden layer *maybe infeasibly large*.
    - Theorem says nothing about the learnability of those parameters.
    - The resulting model *may not generalize to new data*.

### Success or Hype(?) of Deep Learning

- Deep Learning has had tremendous success over the last decade in various domains as we have seen in this course.
- Though the advances are definitely impressive, but it has led to widespread hype in popular *media exaggerating the capabilities of Deep Learning of AI*.
- However, the is not first time that we have had so much interest in AI.
    - There was *increased interest in AI & Neural Networks* and its capabilities in *1950s-1960s abd then again around mid-1980's*.
    - This was followed by bust periods also known as *"AI Winter"* in *mid-1970's and mid-1990's*.

### Neural Networks as Function Approximations

- NN's are excellent and *highly flexible approximators* when they have *lots of training data*.
- However, they *can be fooled and led astray* which has lef to an important area of research on robot NN's.



### Adversarial Attacks on NN'seen

*Adding small but intentionally worst case perturbations* to trainin data to make the NN misclassify the input with high precision.

An active area of research and lots of recent work on "defenses" against such adversarial attacks.

### Neural Network Limitations

- NN's typically require lots of training data and are data hungry (millions of parameters!).
- Some of the deeper NN's can be *computationally very expensive to train.* 
- Susceptible to adversarial attacks.
- Often require *expert knowledge to design and fine tune different architectures!* [Next we will see how we can address this limitation to some extent by Automated Machine Learning (AutoML).]
