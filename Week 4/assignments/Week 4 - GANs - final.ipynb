### Overview

Finally, we will review adversarial machine learning models, a powerful paradigm that encompasses multiple classes of predictive models. These models can be used for anything from image generation to data augmentation to text translation.

### Model Architecture

An adversarial model consists of two pieces: a generator and a discriminator. The generator produces some output using a given noise vector (for example, an artificial image). The goal of the discriminator is then to differentiate real images from generated images.

The discriminator can be trained directly using ground-truth data, and the generator can then be trained in a feedback loop with the discriminator. In this way, the generator aims to fool the discriminator, and the discriminator aims to become more and more discerning.

### Question 1: Generative Models vs Discriminative Models (20 points)

What are the key differences between a generative model and a discriminative model from a statistical point of view? Explain them. 

---

### Answer: ###

A **generative model** learns the **joint probability distribution** of the data, typically written as $(P(X, Y))$ or, in unsupervised settings, just the data distribution $(P(X))$. Because it models how the data are generated, it can be used to **sample new examples**, estimate likelihoods, and in some cases infer class labels through Bayes' rule.

A **discriminative model** learns the **conditional distribution** $(P(Y \mid X))$ directly, or equivalently learns a decision boundary that separates classes. Its goal is not to explain how the data were produced, but to **predict labels from inputs** as accurately as possible. 

The key statistical difference is therefore:
- **Generative:** models the data-generating process.
- **Discriminative:** models the mapping from input to target.

In practice, generative models are more flexible because they can generate synthetic samples and handle missing-data-style reasoning, but discriminative models often achieve better predictive performance on supervised classification tasks because they focus only on the boundary needed for prediction.

### Question 2: GAN Trade-offs (20 points)

What are the main advantages and disadvantages of GANs over standard machine learning models? Explain them.

---

### Answer: ###

**Advantages of GANs:**

1. **High-quality sample generation.** GANs are especially strong at generating realistic images, audio, and other complex data compared with many classical machine learning models.
2. **No explicit likelihood needed.** Unlike some probabilistic generative models, GANs do not require specifying and optimizing a tractable likelihood function.
3. **Useful for data augmentation and simulation.** GANs can create synthetic data to expand small datasets, support training, or help with domain adaptation.
4. **Flexible architecture.** GANs can be adapted for image translation, super-resolution, style transfer, inpainting, and many other tasks.

**Disadvantages of GANs:**

1. **Training instability.** GANs optimize a minimax game between generator and discriminator, which is often harder to train than a standard supervised model.
2. **Mode collapse.** The generator may produce limited varieties of outputs, meaning it learns only part of the true data distribution.
3. **Hard evaluation.** There is no single simple accuracy metric analogous to classification accuracy; sample quality and diversity can be difficult to measure.
4. **Data and compute demands.** GANs often require large datasets, careful tuning, and substantial compute to work well.
5. **Less interpretable than standard models.** Compared with linear models or trees, GAN behavior is harder to explain and debug.

### Question 3: Adversarial Attacks (20 points)

What is an adversarial attack on a machine learning model? Explain how it works.

---

### Answer: ###

An **adversarial attack** is a deliberate manipulation of the input to a machine learning model so that the model makes a wrong prediction, even though the modified input may appear normal to a human.

A machine learning model looks at patterns in the input and uses those patterns to choose an output. An adversarial attack finds the specific parts of the input that influence the model the most, then changes them just enough to confuse the model.

For example, in an image model:

- The model sees a picture of a stop sign.
- An attacker slightly changes some pixel values.
- The picture still looks like a stop sign to a person.
- But the model may now classify it as something else.

The key point is that the changes are not random. They are chosen on purpose to push the model toward the wrong answer.

### Question 4: Retraining (20 points)

A company has found that their internal image recognition tool is producing empirically poor output for a set of images submitted by their users. Upon further investigation, it is determined that these poor performing images are adversarial attacks designed to fool their image recognition software. However, the same set of images are reused by these attackers over and over. Given this set of successful adversarial attacks, how could this be mitigated in the short run? Will this work in the long run? Why or why not?

---

### Answer: ###

Since the attackers are reusing the same adversarial images, one practical mitigation is to **retrain or fine-tune** the model using those attacked examples labeled correctly. This is a form of **adversarial training**.

A reasonable approach would be:
1. Collect the repeated adversarial images.
2. Assign them the correct labels.
3. Add them to the training set, possibly together with normal examples from the same classes.
4. Retrain or fine-tune the model so it learns to classify those inputs correctly.

This would be helpful and **immediate option** for mitigation, as the model has already shown that these examples are weak points. By using them in training, the decision boundary can shift so those known attacks no longer succeed. Because the attack set is reused, the defense can be very effective against this specific threat.

However, this method is not fullproof, as it cannot guarantee robustness to new attacks. Attackers may create slightly modified adversarial examples that still fool the model. So a stronger **long-term defense** would combine retraining with additional safeguards such as broader adversarial training, input filtering/detection, confidence thresholding, and regular model monitoring.

### Question 5: Applications (20 points)

What are potential applications of GANs? Research and briefly present three machine learning innovations that used GANs.

---

### Answer: ###

GANs have many applications because they can learn to generate realistic data. Common uses include image synthesis, super-resolution, image-to-image translation, data augmentation, medical imaging support, video generation, art/style transfer, and anomaly simulation.

Three important machine learning innovations that used GANs are:

### 1. DCGAN (Deep Convolutional GAN)
DCGAN showed that convolutional neural networks could make GAN training much more stable and produce substantially better image samples than earlier fully connected GANs. It helped establish practical design rules for GAN architectures and became a foundational model for later image-generation research.

### 2. pix2pix
pix2pix applied conditional GANs to **image-to-image translation**. Instead of generating an image from pure noise, it learns a mapping from one image representation to another, such as edges to photos, labels to street scenes, or sketches to realistic images. This was important because it showed GANs could solve structured prediction problems, not just unconditional generation.

### 3. CycleGAN
CycleGAN made it possible to translate between two visual domains **without requiring paired training examples**. For example, it can transform horses into zebras or summer scenes into winter scenes using unpaired datasets. Its main innovation was cycle-consistency, which made GAN-based translation much more practical when matched before/after examples were unavailable.
