# Sealed-bid Auction Experiments

- Each student is paired with one robot *(n = 2)*
    - Individual valuation for the object is a random draw  from *u[0, 100]*
    - Each player submits bid
- First-price sealed-bid auctions
    - 5 rounds
    - Robot's strategy: bid = (1/2) value, or *b = v/2*
- Second-price sealed-bid auctions
    - 5 rounds
    - Robot's strategy: bid = value, or *b = v*

## Payoffs

- End of Game
    - highest bidder at the end of the auction is the winner
- Payoff for Winner
    - Payoff = Winner's Value - Price
        - *Price = highest bid in first-price auction*
        - *Price = second highest bid in second-price auction*
- Payoff for Loser
    - Payoff for loser = $0

- Demonstrate how induced-value method works
