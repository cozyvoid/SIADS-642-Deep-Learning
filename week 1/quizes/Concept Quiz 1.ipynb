### 1. What is the main point of induced value theory?

A subject induces value in an experiment to allow the experimenter to gain causal inference.

An experimenter uses subjects’ characteristics to determine the amount of value subjects provide the experiment.

### 3. True or False: The study from the following abstract should be classified as a natural field experiment:

*“This study seeks to estimate the health effects of sanitary drinking water among low-income villages in Guatemala. A random sample of all villages with fewer than 2,000 inhabitants were selected for analysis. Of the 250 villages sampled, 110 were found to have unsanitary drinking water. In these 110 villages, infant mortality rates were, on average, 25 deaths per 1,000 live births, as compared to 5 deaths per 1,000 live births in the 140 villages with sanitary drinking water. Unsanitary drinking water appears to be a major contributor to infant mortality.”*

- False

---
### Why:

This is not a natural field experiment because there is no randomized assignment of treatment. The researchers sampled villages and compared those that already had sanitary or unsanitary water. That makes it an observational study, not an experiment.

### 4. How can randomization help to infer a cause?

It ensures there are little to no systematic differences between treatment and control groups.

---
### Why:

Randomization helps support causal inference because it makes the treatment and control groups comparable on average. Any differences between them are less likely to be systematic, so differences in outcomes can more credibly be attributed to the treatment.

### 7. Select the three characteristics of a natural field experiment.

- Conducted in a real-world setting.
- Subjects are randomized into treatment and control conditions by experimenters.
- Subjects do not know they are in an experiment

---
### Why:

A natural field experiment takes place in a real-world environment, uses random assignment by the researchers, and the subjects are typically unaware they are participating, which helps preserve natural behavior.

### 9. What are nuisance variables?

### 10. In a first-price sealed-bid auction, bidders maximize expected value. When there are only two bidders, the equilibrium bid is ½ of a bidder’s value. As the number of bidder increases, which of the following statements are true?

- The equilibrium bids will increase.
- The auction becomes more competitive.
- The winning bidder’s surplus (or profit) will decrease.

---
### Why:

In a first-price sealed-bid auction, more bidders means stronger competition. Each bidder has a lower chance of winning with a low bid, so they bid more aggressively. That pushes equilibrium bids upward. As bids rise closer to true values, the winner keeps less surplus, so the winner’s profit decreases.
